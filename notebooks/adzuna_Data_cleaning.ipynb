```python
import pandas as pd
df=pd.read_csv("raw_job_data.xls")
```

```python
df.head()
```

```python
df1=df.copy()
```

```python
df1.drop(columns=["longitude","latitude"],inplace=True)
```

```python
df1.rename(columns={'location.display_name': 'Location'},inplace=True)
```

```python
df1['salary_max']= pd.to_numeric(df1['salary_max'])
```

```python
df1.drop(columns="category.tag",inplace=True)
```

```python
df1.rename(columns= {"category.label" : "Category"},inplace=True)
```

```python
known_cities = [
    'Navi Mumbai', 'Mumbai', 'Bangalore', 'Hyderabad', 'Gurgaon', 'Noida',
    'Chennai', 'Pune', 'Kolkata', 'Ahmedabad', 'Jaipur', 'Surat', 'Vadodara',
    'Nagpur', 'Coimbatore', 'Bhopal', 'Kota', 'Ajmer', 'Mangalore',
    'Thiruvananthapuram', 'Delhi'
]

def cities(Location):
    if pd.isna(Location):
        return 'unknown'
    for city in known_cities:
        if city.lower() in Location.lower():
            return city
    return 'other/state only'  

df1['city']=df1['Location'].apply(cities)
```

```python
df1.drop(columns=["__CLASS__","redirect_url","adref","location.__CLASS__","location.area","company.__CLASS__","category.__CLASS__"],inplace=True)
```

```python
df1["created"]=pd.to_datetime(df1["created"])
```

```python
df1.rename(columns={"company.display_name":"company","salary_is_predicted":"salary_predicted"},inplace=True)
```

```python
skills_req=['python','power bi','power-bi','powerbi','sql','excel','tableau']

def fetch_skill(Descrip):
    if pd.isna(Descrip):
        return 'unknown'
    for skill in skills_req:
        if skill.lower() in Descrip.lower():
            return skill
    return 

df1['Skills']=df1['description'].apply(fetch_skill)
```

```python
# Combine title and description for searching (fill NaN with empty string first to avoid errors)
df1['search_text'] = (df1['title'].fillna('') + ' ' + df1['description'].fillna('')).str.lower()

# Create skill flag columns
df1['has_python'] = df1['search_text'].str.contains('python', case=False)
df1['has_sql'] = df1['search_text'].str.contains('sql', case=False)
df1['has_excel'] = df1['search_text'].str.contains('excel', case=False)
df1['has_powerbi'] = df1['search_text'].str.contains('power bi|powerbi|power-bi', case=False)
df1['has_tableau'] = df1['search_text'].str.contains('tableau', case=False)

# Check how many postings mention each skill
skills = ['has_python', 'has_sql', 'has_excel', 'has_powerbi', 'has_tableau']
print(df1[skills].sum())
```

```python
df1.drop(columns='search_text',inplace = True)
```
