```python
import pandas as pd
df=pd.read_csv("postings.csv")
```

```python
df2=df.copy()
```

```python
df2=df2[df2['job_title'].str.contains('data analyst',case=False,na=False)]
```

```python
df2.drop(columns='job_link',inplace=True)
```

```python
df2=df2.dropna(subset='job_summary')
```

```python
df2.count()
```

```python
df2['job_skills']=df2['job_skills'].fillna("Not Specified")
```

```python
def parse_location(loc):
    if pd.isna(loc):
        return pd.Series(['Unknown', 'Unknown'])
    parts = [p.strip() for p in loc.split(',')]
    if len(parts) == 1:
        # Just a country, e.g. "United States"
        return pd.Series(['Unknown', parts[0]])
    elif len(parts) == 2:
        # "City, ST" - US format
        return pd.Series([parts[0], 'United States'])
    else:
        # "City, Region, Country" - international
        return pd.Series([parts[0], parts[-1]])

df2[['city', 'country']] = df2['job_location'].apply(parse_location)
```

```python
skills_to_check = ['python', 'sql', 'excel', 'power bi', 'tableau']

for skill in skills_to_check:
    col_name = 'has_' + skill.replace(' ', '_')
    df2[col_name] = df2['job_skills'].str.lower().str.contains(skill, na=False)
```

```python
df2.drop(columns='job_location',inplace=True)
```

```python
df2.reset_index(drop=True,inplace=True)
```
